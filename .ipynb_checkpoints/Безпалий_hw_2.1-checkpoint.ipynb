```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C")
```

```python
print(df[8])
df = df[8]
```

```python
#Завдання 1
df.head()
```

```python
#Завдання 2
df.shape
```

```python
#Завдання 3
df = df.replace(to_replace='—', value=np.nan, regex=True)
df
```

```python
#Завдання 4
df.dtypes
```

```python
#Завдання 5
df[['1950', '1960', '1970', '2014', '2019']] = df[['1950', '1960', '1970', '2014', '2019']].astype('float')
df.dtypes
```

```python
#Завдання 6
pd.isnull(df).sum()
```

```python
#Завдання 7
df = df.drop(labels=[len(df)-1], axis=0)
df
```

```python
#Завдання 8
d = list(df)[1:]
for i in d:
    df2 = df[i].mean()
    df[i] = df[i].fillna(df2).round()
df
```

```python
#Завдання 9
d = list(df["Регіон"])
val = list(df["2019"])
mean = round(sum(val)/len(val))
# df.index[df['2019'] > mean ].tolist()
df[df["2019"]>mean]["Регіон"].tolist()
```

```python
#Завдання 10
df.loc[df["2014"] == df["2014"].max()]["Регіон"]
```

```python
#Завдання 11
plt.barh(df["Регіон"], df['2019'])
```

```python
# Лабораторна 4_1
# Завдання 1
# Графік 1
years = list(df)[1:]
reg = list(df["Регіон"])
a = df[df["Регіон"] == "Крим"].values.tolist()[0][1:]
plt.plot(years,a)
```

```python
# Графік 2
index = years
values = a
plt.bar(index,values)
plt.legend([reg[0]], loc=1)
plt.show()
```

```python
# Графік 3
data = {'Київ':df[df["Регіон"] == "Київ"].values.tolist()[0][1:],
       'Чернігівська':df[df["Регіон"] == "Чернігівська"].values.tolist()[0][1:],
       'Закарпатська':df[df["Регіон"] == "Закарпатська"].values.tolist()[0][1:]}
df1 = pd.DataFrame(data)
df1.plot(kind='bar',stacked=True)
plt.show()
```
