```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('C:\Files\My projects by Python\GeneratorDataBase\Base2.csv', encoding='cp1251')
sales
```

```python
sns.set(style="darkgrid")       #style the plot background to become a grid
genderCount  = sns.countplot(x="Пол покупателя", data =sales).set_title("Количество женщин и мужчин")
```

```python
sales['Время'] = pd.to_datetime(sales['Время'])
sales['Часы'] = (sales['Время']).dt.hour
```

```python
sales['Дата'] = pd.to_datetime(sales['Дата'])
sales['День'] = (sales['Дата']).dt.day
sales['Месяц'] = (sales['Дата']).dt.month
sales['Год'] = (sales['Дата']).dt.year
```

```python
genderCount  = sns.lineplot(x="Часы",  y = 'Количество',data =sales).set_title("Продажи продукта в час")
```

```python
genderCount  = sns.relplot(x="Часы",  y = 'Количество', col= 'Месяц' , row= 'Отдел', kind="line", hue="Пол покупателя", style="Пол покупателя", data =sales)
```

```python
genderCount  = sns.relplot(x="Часы",  y = 'Общая сумма', col= 'Месяц' , row= 'Отдел', estimator = None, kind="line", data =sales)
```

```python
sns.boxenplot(y = 'Линейка продукта', x = 'Количество', data=sales )
```

```python
sns.countplot(y = 'Линейка продукта', data=sales, order = sales['Линейка продукта'].value_counts().index )
```

```python
sns.boxenplot(y = 'Линейка продукта', x = 'Общая сумма', data=sales )
```

```python
sns.stripplot(y = 'Линейка продукта', x = 'Общая сумма', hue = 'Пол покупателя', data=sales )
```

```python
sns.countplot(x="Тип оплаты", data =sales).set_title("Рейтинг типа оплаты") 
```
