```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Лабораторна 4_1
# Завдання 3
def draw_sierpinski(x, y, length, depth, max_depth):
    if depth == max_depth:
        return
    
    # Розбиваємо трикутник на 3 нових трикутники
    x1, y1 = x, y
    x2, y2 = x + length, y
    x3, y3 = x + length / 2, y + length * (3 ** 0.5) / 2

    # Малюємо 3 нових трикутники
    plt.scatter([x1, x2, x3], [y1, y2, y3], s=1, color='black')
    
    # Рекурсивно викликаємо функцію для кожного з нових трикутників
    draw_sierpinski(x1, y1, length / 2, depth + 1, max_depth)
    draw_sierpinski(x2, y2, length / 2, depth + 1, max_depth)
    draw_sierpinski(x3, y3, length / 2, depth + 1, max_depth)
```

```python
# Створюємо новий графік
plt.figure(figsize=(10, 10))

# Встановлюємо масштаб по осях x та y рівним
plt.axis('equal')

# Викликаємо функцію для малювання фракталу
draw_sierpinski(0, 0, 100, 0, 7)

# Відображаємо графік
plt.show()
```

```python
# Завдання 4
df = pd.DataFrame({'Момент спостереження':[1,2,3,4,5,6],
                 'Доходність А,%':[25,-10,10,5,35,13],
                 'Доходність B,%':[0,15,-5,5,20,25],
                 'Доходність C,%':[10,25,-15,-5,-5,15]})
df
```

```python
# График залежності прибутковості А від прибутковості B
sns.regplot(x='Доходність B,%', y='Доходність А,%', data=df)
plt.title('Залежність прибутковості А від прибутковості B')
plt.show()

# График залежності прибутковості А від прибутковості С
sns.regplot(x='Доходність C,%', y='Доходність А,%', data=df)
plt.title('Залежність прибутковості А від прибутковості С')
plt.show()

# График залежності прибутковості В від прибутковості С
sns.regplot(x='Доходність C,%', y='Доходність B,%', data=df)
plt.title('Залежність прибутковості В від прибутковості С')
plt.show()
```

```python
# Знаходимо пару акцій, які найменше залежать за прибутковістю один від одного.
```

```python
# обчислюємо матрицю кореляцій
corr_matrix = df.corr(method='pearson')

# знаходимо найменше значення кореляції та відповідні назви стовпців
min_corr_value = corr_matrix.min().min()
min_corr_cols = corr_matrix[corr_matrix == min_corr_value].stack().index.tolist()

# виводимо результат
print(f'Найменша кореляція ({min_corr_value}) між акціями {min_corr_cols[0][0]} та {min_corr_cols[0][1]}')

# Висновок
# Чим ближче коефіцієнт кореляції Пірсона до 1 або -1, тим сильніша лінійна залежність між парами стовпців. 
# Чим ближче до 0, тим слабша або відсутня лінійна залежність.
# Таким чином, акції А і С найменше залежать за доходністю один від одного.
```
