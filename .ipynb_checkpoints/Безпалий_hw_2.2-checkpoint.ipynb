```python
#Завадння 5
null_list = pd.isnull(data).sum().tolist()
print(null_list)
```

```python
#Завдання 6
d = data.columns.to_list() #Отримуємо список назв стовпців
names = [] # Список з назвами стовпців, у котрих присутні пропуски
for i in null_list: #Заповнюємо список вікористовуючи список з завадння №5
    if i != 0:
        names.append(d[null_list.index(i)])
print(names)
names.pop(names.index('Язык.программирования')) #Прибираємо виключення зі списку
print(names)

data = data.drop(columns=names) #Прибираємо стовпці по назвах
data
```

```python
#Завдання 7
null_list2 = pd.isnull(data).sum().tolist()
print(null_list2)

d = data.columns.to_list() #Отримуємо список назв стовпців
names2 = [] # Список з назвами стовпців, у котрих присутні пропуски
for i in null_list2: #Заповнюємо список вікористовуючи список null_list2
    if i != 0:
        names2.append(d[null_list2.index(i)])
print(names2)
```

```python
#Завдання 10
python_data = data[data["Язык.программирования"] == "Python"]
python_data
```

```python
#Завдання 11
python_data.shape
```

```python
#Завдання 12
python_data_2 = python_data.groupby("Должность")
python_data_2.first()
```

```python
#Завдання 13
python_data_2.agg({'Зарплата.в.месяц' : ['max', 'min']})
```

```python
#Завдання 14
#Створюємо список з назвами посад
d = python_data_2["Должность"].first().tolist()
print(d)

#Створюємо функцію, котра повертає середню з.п. відповідно вказаній посаді
def fill_avg_salary(n):
    return round(python_data[python_data["Должность"] == n]["Зарплата.в.месяц"].mean())

#Створюємо список у котрий додамо середні з. п. відповідно до посад
mean_salary = []
for i in d:
    mean_salary.append(fill_avg_salary(i))
print(mean_salary)
python_data

#Створимо новий стовпець "avg"
python_data["avg"] == ""

#Заповнимо стовпець "avg" середніми з.п відповідно до посади
for i in d:
    python_data.loc[python_data["Должность"] == i, ["avg"]] = mean_salary[d.index(i)]
    
#Знову згрупуємо датафрейм
python_data_2 = python_data.groupby("Должность")

#Виводимо
python_data_2.first()
```

```python
# Лабораторна 4_1
# Завдання 1
# Графік 1
zp = list(python_data_2["Зарплата.в.месяц"])
zp
job = []
for i in range(0, len(zp)):
    job.append(zp[i][0])
print(job)
mean_s = []
for i in job:
    mean_s.append(round(python_data.loc[python_data['Должность'] == i, 'Зарплата.в.месяц'].mean(),2))
print(mean_s)


labels = job
values = mean_s
colors = ['yellow','darkgreen','red','blue','orange']
plt.pie(values,labels=labels,colors=colors)
plt.axis('equal')
plt.legend(["Середня зарплата в місяць"], loc=3)
plt.show()
```

```python
# Графік 2
mean_age = []
mean_exp = []
for i in job:
    mean_age.append(python_data.loc[python_data['Должность'] == i, 'Возраст'].mean())
    mean_exp.append(python_data.loc[python_data['Должность'] == i, 'exp'].mean())

index = np.arange(5)
data = {'Сер. вік': mean_age,
        'Сер. заг. досвід роб.': mean_exp}
df = pd.DataFrame(data)
df.plot(kind='bar')
plt.show()
```
