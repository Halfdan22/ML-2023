```python
#Завдання 6
d = data.columns.to_list() #Отримуємо список назв стовпців
names = [] # Список з назвами стовпців, у котрих присутні пропуски
for i in null_list: #Заповнюємо список вікористовуючи список з завадння №5
    if i != 0:
        names.append(d[null_list.index(i)])
print(names)
names.pop(names.index('Язык.программирования')) #Прибираємо виключення зі списку
print(names)

data = data.drop(columns=names) #Прибираємо стовпці по назвах
data
```

```python
#Завдання 14
#Створюємо список з назвами посад
d = python_data_2["Должность"].first().tolist()
print(d)

#Створюємо функцію, котра повертає середню з.п. відповідно вказаній посаді
def fill_avg_salary(n):
    return round(python_data[python_data["Должность"] == n]["Зарплата.в.месяц"].mean())

#Створюємо список у котрий додамо середні з. п. відповідно до посад
mean_salary = []
for i in d:
    mean_salary.append(fill_avg_salary(i))
print(mean_salary)

#Створимо новий стовпець "avg"
python_data["avg"] == ' '

#Заповнимо стовпець "avg" середніми з.п відповідно до посади
for i in d:
    python_data.loc[python_data["Должность"] == i, ["avg"]] = mean_salary[d.index(i)]
    
#Знову згрупуємо датафрейм
python_data_2 = python_data.groupby("Должность")

#Виводимо
python_data_2.first()
```
